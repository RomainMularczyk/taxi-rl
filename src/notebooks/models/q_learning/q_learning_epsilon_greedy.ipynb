```python
import os; os.chdir("../../..")
print(os.getcwd())
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from lib.evals.models.metrics import aggregate_metrics, graph_training_evolution
from lib.formulas.q_learning import QLearning
from lib.policies.EpsilonGreedyPolicy import EpsilonGreedyPolicy
```

```python
SEED = np.random.randint(500)
TRAINING_EPISODES = 2000
```

In this experiment we will compare how the **epsilon greedy policy** perform both on :
- Varying environment initialization
- Same environment initialization (the same seed is used)

# First experiment (Seeded)

## Hyperparameters

**Q-Learning :**
* Advantage: False
* Gamma: 0.9
* Learning rate: 0.01

**Policy :**
* Epsilon: 0.9
* Decay rate: 0.0005
* Legal: True

```python
env = gym.make("Taxi-v3", render_mode="ansi")
policy = EpsilonGreedyPolicy(
    game_env=env,
    epsilon=0.9,
    decay_rate=0.05,
    legal=True,
    seed=SEED
)
q_learning = QLearning(
    policy=policy,
    observation_space=env.observation_space.n,
    action_space=env.action_space.n,
    advantage=False,
    max_steps=200,
    gamma=0.7,
    lr=0.1
)
q_learning_metrics = q_learning.train(TRAINING_EPISODES)
print(aggregate_metrics(q_learning_metrics))
graph_training_evolution(q_learning_metrics)
```
